```python
# First we need to define the Node class.
# The class can be initiated with the value parameter, if the argument is not provided it will be set to None
# The children of the node are also assigned initially with the value of None
class TreeNode:
    def __init__(self, value=None):
        self.left = None  # left child node
        self.right = None # right child node
        self.node_value = value # node value
        
    
        
```

```python
# Next we define the Tree class 
# When initiated it is an empty tree with the root_node set to None
class Tree:
    def __init__(self):
        self.root_node = None

    # To allow new values to be added to the tree class in a form of nodes, the insert function is defined.
    # It accept one argument (new_value)
    def insert(self, new_value):
       
       
        # And now we run through numerous condition checks
        # Fist, we check if the root node exist, if root is None, we create new node as the root of
        # the tree and return from the insert function.
        # Relevant information is printed to the user.
        if self.root_node is None:
            
            self.root_node = TreeNode(new_value)
            print(new_value, ' - Inserted as root')
            return
        # If root node exist, we execute a private method called insertRecursive providing the existing root 
        # and new_value as an argument.
        else:
            self._insertRecursive(self.root_node, new_value)
     
    # This is a private recursive method to insert new values into the tree at specific position which is going
    # to be determined at the insertion time. 
    # To better understand the insert process, relevant lines of the code are printing information, allowing to 
    # draw the tree if necessary.
    
    def _insertRecursive(self, node, new_value):
        # compering new_value with the node_value of the current node
        if new_value < node.node_value:
            if node.left is None:
                # if value is less than current node value and the left child does not exist for current node
                node.left = TreeNode(new_value)
                # New left-node is created and added as a left child and relevant information printed
                print(new_value," Inserted on left of ",node.node_value)
                return
            else:
                # Otherwise (the left child of a current node already exist), we call recursive the method and pass 
                # on the existing left child as the current node
                self._insertRecursive (node.left, new_value)
                return
        else:
            if node.right is None:
                # if value is greater than current node value and the right child does not exist for current node
                node.right = TreeNode(new_value)
                # New right node is created and added as a right chiled and relevant information printed
                print(new_value," Inserted on right of ",node.node_value)
                return
            else:
                # Otherwise (the right child of a current node already exist), we call recursive 
                # the method and pass on the existing right child as the current node
                self._insertRecursive (node.right, new_value)
                return
             
```

```python
# Test input: [9,30,36,1,40,25,33]

tree = Tree()
tree.insert(9)
tree.insert(30)
tree.insert(36)
tree.insert(1)
tree.insert(40)
tree.insert(25)
tree.insert(33)





```
